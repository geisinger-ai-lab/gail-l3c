## Demonstrating Hive SQL query on spark dataframe...

```python
import pyspark.sql.functions as f
from pyspark.sql import SparkSession

from src import common
```

```python
spark = common.get_spark_session()

visit_occurrence = spark.read.csv('../data/raw_sample/training/visit_occurrence.csv', header=True)
visit_occurrence.show()
```

Create/replace a temp view. Then use `spark.sql` function to query that view and execute SQL on DataFrame...

```python
visit_occurrence.createOrReplaceTempView("visit_occurrence")

new_df = spark.sql("select person_id,visit_start_date from visit_occurrence limit 3")
new_df.show()
```
